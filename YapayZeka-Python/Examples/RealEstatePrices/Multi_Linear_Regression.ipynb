```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
import pickle
```

```python
# Read the dataset from the CSV file and load it into a DataFrame
data = pd.read_csv("multilinearregression.csv", sep=";")
```

```python
# Print the dataset
print(data)
```

```python
# Create and train the linear regression model
reg = linear_model.LinearRegression()
reg.fit(data[["alan", "odasayisi", "binayasi"]], data["fiyat"])
```

```python
# Predict the price for new data points and print the results
print(reg.predict([[230, 4, 10]])) 
print(reg.predict([[230, 6, 0]])) 
print(reg.predict([[355, 3, 20]])) 
```

```python
pickle.dump(reg, open("predticton.pickle", "wb"))
```

```python
print(reg.coef_)      
print(reg.intercept_) 
```

```python
a = reg.intercept_
b1 = reg.coef_[0]
b2 = reg.coef_[1]
b3 = reg.coef_[2]
```

```python
x1 = 230
x2 = 4
x3 = 10
y = a + b1 * x1 + b2 * x2 + b3 * x3
print("El ile hesaplanan y değeri:", y)
```
